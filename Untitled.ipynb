```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel("Excel prueba.xlsx")
df
```

```python
# Ingresar producto de interés acá
producto = "BANDEJA ESCRITORIO METAL TRIPLE REJILLA BLANCO"
```

```python
palabras = producto.split(" ")

df["Coincidencias"] = 0
for palabra in palabras:
    df["Coincidencias"] += df["Producto"].str.contains(palabra)
    
df
```

```python
df2 = df.groupby("Taxonomía").mean()
df2 = df2.sort_values(by = ["Coincidencias"], ascending = False)
df2 = df2.rename(columns = {"Coincidencias": "Tasa de coincidencia"})
df2
```

```python
# Mostrar las primeras 5 recomendaciones

print("Recomendaciones:")
head = df2.head()
i = 1
for tax, t in zip(head.index, head["Tasa de coincidencia"]):
    print(f"{i}. {tax} (tasa: {t})")
    i += 1
```

```python
# Escogemos la primera recomendación
recomendacion = df2.index[0]

df = df.drop("Coincidencias", axis=1)
df.loc[len(df.index)] = [producto, recomendacion]
df
```
